#### Exercise 2.1 Byte pair encoding of unknown words 

```python
import tiktoken 

tokenizer = tiktoken.get_encoding("gpt2")

text = "Akwirw ier"

# encoding 
integer = tokenizer.encode(text)
print(f"encoding: ",integer)

# decoding 
string = tokenizer.decode(integer)
print("string: ",string)
```
